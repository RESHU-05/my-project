## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='black',linestyle='dashdot')
plt.xlabel('X',size=15)
plt.ylabel('Y',size=15)
plt.title('X-Y sine wave',size=15)
plt.show()
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
stores=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
clothing=np.array([82,78,48,46,39])
equipment=np.array([48,31,51,29,18])
accessories=np.array([15,20,14,13,9])
var=[x for x, _ in enumerate(stores)]
plt.bar(var,clothing,width=0.8,label='clothing',color='cyan',bottom=equipment+accessories)
plt.bar(var,equipment,width=0.8,label='equipment',color='magenta',bottom=accessories)
plt.bar(var,accessories,width=0.8,label='accessories',color='blue')
plt.ylabel("numbers",size=15)
plt.xlabel("places",size=15)
plt.xticks(var,stores,rotation=90)
plt.title("Stores",size=15)
plt.grid(False)
plt.legend(loc='upper right')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
d=sns.load_dataset('titanic')
d.head(5)
```

```python
sns.barplot(x='pclass',y='survived',hue='class',data=d,palette='summer',ci=0,order=[3,2,1],saturation=0.80,dodge=False)
plt.xlabel("passenger class",size=15)
plt.ylabel("survival of passenger",size=15)
plt.title('survival of passenger against passenger class',size=15)
plt.legend(loc='upper left')
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
da=sns.load_dataset('tips')
da.head(5)
sns.distplot(da['total_bill'],color='green',vertical=True)
plt.xlabel('total_bill',size=15)
plt.ylabel('data',size=15)
plt.title('distribution',size=15)
plt.show()
```

# Q.7
- Also plot the KDE Graph for the above data

```python
sns.kdeplot(da['total_bill'])
plt.xlabel('data',size=15)
plt.ylabel('total_bill',size=15)
plt.title('distribution',size=15)
plt.show()
```

## Q.9 

- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
country=['United States','Great Britain','India','Russia','Germany']
medal=['46','27','26','19','17']
color=['red','yellow','green','orange','blue']
explode=(0,0,0.1,0,0)
plt.pie(medal,labels=country,colors=color,autopct='%1.1f%%',explode=explode,pctdistance=0.5,shadow=True,startangle=90)
plt.title('country',size=15,color='purple')
plt.show()
```
